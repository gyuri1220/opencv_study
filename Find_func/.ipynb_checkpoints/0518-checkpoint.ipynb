```python
#원본 -> 가우시안블러 이미지 생성 & 저장
import cv2

# 1. 이미지 읽기
img = cv2.imread('image.png')

# 2. 가우시안 블러 적용
# (5, 5)는 커널 사이즈 (홀수), 0은 표준 편차
blur_img = cv2.GaussianBlur(img, (5, 5), 0)

# 3. 결과 확인
cv2.imshow('Original', img)
cv2.imshow('Blurred', blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('Blurred.png', blur_img)
```

```python
#원본 -> 임펄스잡음 이미지 생성 & 저장
import cv2
import numpy as np
import random

def add_impulse_noise(image_path, prob):
    # 1. 이미지 읽기 (입력받은 image_path 사용)
    img = cv2.imread(image_path)
    
    # 변수 이름을 img로 통일합니다.
    if img is None:
        print("이미지를 불러올 수 없습니다. 경로를 확인하세요.")
        return None

    noisy_image = img.copy()
    
    # 2. 잡음이 적용될 픽셀 수 계산 (img.size 사용)
    num_noise = np.ceil(prob * img.size / 3) 
    
    for _ in range(int(num_noise)):
        # img.shape 사용
        y = np.random.randint(0, img.shape[0])
        x = np.random.randint(0, img.shape[1])
        
        # 3. Salt and Pepper 잡음 적용
        if np.random.rand() < 0.5:
            noisy_image[y, x] = [255, 255, 255] # 흰색 점 (Salt)
        else:
            noisy_image[y, x] = [0, 0, 0]       # 검은색 점 (Pepper)
            
    return noisy_image

# ========================================================
# 실제로 함수를 사용하고 이미지를 저장하는 부분입니다.
# ========================================================

# 1. 원본 이미지 불러오기 (화면 표시용)
origin_img = cv2.imread('image.png')

# 2. 함수를 실행해서 잡음 섞인 이미지를 변수(result_image)에 받아오기
# (확률을 0.05로 주면 전체 픽셀의 약 5%에 잡음이 낍니다)
result_image = add_impulse_noise('image.png', prob=0.5)

# 3. 결과가 정상적일 때만 화면에 보여주고 저장하기
if result_image is not None:
    cv2.imshow('Original', origin_img)
    cv2.imshow('Noisy Image', result_image)
    
    cv2.waitKey(0) #키보드 눌러야 닫힘
    cv2.destroyAllWindows()
    
    # 이 코드가 실행되면 파이썬 파일과 같은 폴더에 저장됩니다!
    cv2.imwrite('noisy_image2.png', result_image)
    print("성공적으로 'noisy_image.png' 가 저장되었습니다.")
```

```python
#이미지 샤프닝(선명)
# 에지 근방에서 픽셀 값의 명암비가 커지도록 조정
#샤프닝은 블러링된 영상을 사용해야함 / 블러링이 적용된 영상(언샤프)을 이용하여 날카로운 영상을 생성하는 필터를 언샤프 마스크 필터

import cv2
import numpy as np

# 1. 이미지 불러오기 (그레이스케일 또는 컬러)
img = cv2.imread('circle.jpg')

# 2. 가우시안 블러를 이용한 부드러운 이미지 생성
# 커널 사이즈(ksize)가 클수록 더 강한 샤프닝 효과가 발생합니다.
blurred = cv2.GaussianBlur(img, (0, 0), 3)

# 3. 언샤프 마스크(Unsharp Mask) 적용
# img + (img - blurred) * alpha (alpha는 가중치)
sharpened = cv2.addWeighted(img, 2.0, blurred, -1.0, 0)

# 결과 출력
cv2.imshow('Original', img)
cv2.imshow('Sharpened', sharpened)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite('Sharpened.jpg', sharpened)
```
